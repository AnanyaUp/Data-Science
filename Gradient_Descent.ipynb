```python
# Task - Design and Gradient Descent Algorithm
# Step 1 : y = 3x + 0.7
# Step 2 : Design Gradient Descent Function

import numpy as np
import pandas as pd

import warnings
warnings.filterwarnings('ignore')
```

```python
# Learning Rate => alpha = 0.01

# Step A) Generate a data of size = 30 inrange of 1-50
x = np.random.randint(1,50,30)
y = 3*x + 0.7

# Step B) Convert this into a dataframe
df = pd.DataFrame({'x':x,'y':y})
df.head()
```

```python
# Step C) C (1): Let a hypothetical equation be y' = Wx + b -> Let W = 5 and b = 1
#          C (2): Make a function to calculate y' and append it to the df

W0 = 5
b0 = 1
y_pred = W0*x + b0
df['y_pred'] = y_pred
df.head()
```

```python
# Step D) Calculating Errors = y-y_pred
df['error'] = df['y'] - df['y_pred']
df.head()
```

```python
df.drop(['y_pred','error'],axis=1,inplace=True)
df.head()
```

```python
def update(w, b, x, y, al):
    # Calculate y_cal (predicted value)
    y_cal = w * x + b
    # Calculate the error
    E = y - y_cal
    # Update w and b using gradient descent
    w = w - (al * E * x)
    b = b - (al * E)
    return w, b

wo = 5
bo = 1
print("Old value of wo and bo: ", wo, bo)

for i in range(30):
    wo, bo = update(wo, bo, df['x'][i], df['y'][i], 0.00001)
    print("New value of wo and bo: ", wo, bo)

print("New value of wo and bo: ", wo, bo)
```

```python
# Chat GPT - Method :
import numpy as np
import random

# Step 1: Generate a random sample of 30 values in the range 1-40
random.seed(42)  # for reproducibility
x = np.array(random.sample(range(1, 41), 30))

# Step 2: Calculate y = 3*x + 0.7
y = 3 * x + 0.7

# Step 3: Initialize W0 and b0
W0 = 5  # initial weight
b0 = 1  # initial bias

# Learning rate
lr = 0.00001

# Function to compute the gradient descent updates in a single epoch
def gradient_descent_single_epoch(x, y, W0, b0, lr):
    n = len(x)
    for i in range(n):
        # Predicted y
        y_pred = W0 * x[i] + b0

        # Update W0 and b0 using the provided formulas
        W0 = W0 - lr * (y[i] - y_pred) * x[i]
        b0 = b0 - lr * (y[i] - y_pred)

    return W0, b0

# Step 4: Update the values of W0 and b0 using gradient descent for a single epoch
W0_updated, b0_updated = gradient_descent_single_epoch(x, y, W0, b0, lr)

# Step 5: New equation y_n = W0 * x + b0
print(f"Updated equation after a single epoch: y_n = {W0_updated:.2f} * x + {b0_updated:.2f}")
```
